```python
df = pd.read_csv('Resources/model2_clean_data.csv')
df
```

```python
#Divide the target column and the features column
y = df['stroke']
X = df.drop(['stroke'], axis = 1)
```

```python
# Split the data in training and testing using sklearn
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.8, random_state=12)
```

```python
#Scale the data using StandardScaler
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Defien test_model function that will test multiple models.
def test_model(model, data):
    X_train_scaled, X_test_scaled, y_train, y_test = data
    reg = model.fit(X_train_scaled, y_train)
    oversample = RandomOverSampler(sampling_strategy=0.5)
    X_over, y_over = oversample.fit_resample(X_train_scaled, y_train)
    steps = [('over', RandomOverSampler(sampling_strategy=0.5)), 
         ('model', model.fit(X_over, y_over))]
    pipeline = Pipeline(steps=steps)
    
    print(f'Model: {type(reg).__name__}')
    print(f'Train score: {pipeline.score(X_train_scaled, y_train)}')
    print(f'Test Score: {pipeline.score(X_test_scaled, y_test)}\n')
    y_pred = pipeline.predict(X_train_scaled)
    y_true = y_test
    print(confusion_matrix(y_true, y_pred))
    print(classification_report(y_test, y_pred))
       
```
